```python
#Scipt to move files around

import os
import shutil
import csv

# array of files to transport
files = []
with open("trainfile.csv", 'r') as f:
    reader = csv.reader(f, delimiter=',')
    for row in reader:
        if row[0] != "Id":
            files.append(row[0] + "_green.png")
            files.append(row[0] + "_yellow.png")
            files.append(row[0] + "_blue.png")
            files.append(row[0] + "_red.png")
print(files)

```
